```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from urllib.request import urlretrieve
```

```python
urlretrieve('https://archive.ics.uci.edu/ml/machine-learning-databases/00477/Real%20estate%20valuation%20data%20set.xlsx',
           "Real_Estate.xlsx")
```

```python
Dataframe = pd.read_excel("Real_Estate.xlsx")
```

```python
Dataframe.describe()
```

```python
Dataframe.head()
```

```python
Dataframe['X1 transaction date'] = pd.to_datetime(Dataframe['X1 transaction date'])
```

```python
plt.scatter(Dataframe['X5 latitude'],Dataframe['X6 longitude'], alpha=.5 ,marker="1",vmax=125, vmin=120)
```

```python
plt.plot(Dataframe['Y house price of unit area'],Dataframe['X4 number of convenience stores'], 'o' , ms =20, mec = "g", mfc="r") #marker|line|color
```

```python
font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 12,
        }
```

```python
plt.plot(Dataframe['Y house price of unit area'],Dataframe['X2 house age'], '*' , c='#34495E' )
# plt.plot(Dataframe['Y house price of unit area'],Dataframe['X3 distance to the nearest MRT station'])
plt.xlabel("House price of unit area", fontdict=font)
plt.ylabel("House age in yrs", fontdict=font)
```
